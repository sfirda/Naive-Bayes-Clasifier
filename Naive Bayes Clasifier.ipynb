```python
import numpy as np
import pandas as pd
```

```python
iris=pd.read_csv("iris.csv")
iris
```

```python
iris.head()
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X=iris.values[:,0:4]
Y=iris.values[:,4]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
a=pd.DataFrame(X_test)
a
```

```python
clf_gini=tree.DecisionTreeClassifier(criterion="gini",max_depth=None,min_samples_leaf=2)
clf_gini=clf_gini.fit(X_train,y_train)
```

```python
clf_entropy=tree.DecisionTreeClassifier(criterion="entropy",max_depth=None,min_samples_leaf=2)
clf_entropy=clf_entropy.fit(X_train,y_train)
```

```python
y_pred_gini=clf_gini.predict(X_test)
y_pred_gini
```

```python
y_pred_info=clf_entropy.predict(X_test)
y_pred_info
```

```python
accuracy_score(y_test,y_pred_gini)*100
```

```python
accuracy_score(y_test,y_pred_info)*100
```

```python
iris_pred_gini=pd.DataFrame(y_pred_gini)
iris_pred_gini.to_csv("iris_pred_gini.csv")
iris_pred_info=pd.DataFrame(y_pred_info)
iris_pred_info.to_csv("iris_pred_info.csv")
```

```python
ftrs=['sepal length','sepal width','petal length','petal width']
cls=['viginica','sentosa','versicolor']
import matplotlib.pyplot as plt
fig, axes=plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=300)
tree.plot_tree(clf_gini,feature_names=ftrs,
               class_names=cls,filled=True)
fig.savefig('iristregini.png')
```

```python
ftrs=['sepal length','sepal width','petal length','petal width']
cls=['viginica','sentosa','versicolor']
import matplotlib.pyplot as plt
fig, axes=plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=300)
tree.plot_tree(clf_entropy,feature_names=ftrs,
               class_names=cls,filled=True)
fig.savefig('iristreeinfo.png')
```
